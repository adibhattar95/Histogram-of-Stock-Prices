## Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style('darkgrid')
sns.set_color_codes(palette = 'bright')
import warnings
warnings.filterwarnings("ignore")
```

## Load and view the dataset

```python
df = pd.read_csv('interviewAssignment.csv')
pd.set_option('display.max_columns', 100)
df.head(10)
```

## Initial Cleaning of the Data

```python
#Take a subset of df of columns required
df_new = df.iloc[:, :12]
df_new = df_new.drop(['fcfNtm'], 1)
df_new = df_new[df_new['ebitdaNtm'] > 0]
df_new = df_new[df_new['salesNtm'] > 0]
df_new = df_new[df_new['price'] <= 1000]

#Check for null values
print(df_new.isnull().sum())
print('='*50)

#Drop null values
df_new = df_new.dropna()

#Add column for Price Changes
df_new['Price Change'] = df_new['price'].diff(periods = -1).abs()

#Last row will have NA value, replace with 0
df_new = df_new.fillna(0)

#Add column for Price Change Intervals
df_new['Price Change Interval'] = pd.cut(x = df_new['Price Change'], bins = [0, 25, 50, 75, 100, 125, 150, 175, 200, np.inf],
                                         labels = ['0 - 25', '25 - 50', '50 - 75', '75 - 100', '100 - 125', '125 - 150', 
                                                  '150 - 175', '175 - 200', '>200'])
df_new.head()
```

## Define function to map month to timeframe

```python
def timeframe(dataframe, date_column, month_year_column, new_column):
    
    """
    Function to determine timeframe in months from date column provided
    First extract month and year from date column
    Get a list of unique month and year values
    Assign each month/year combination a timeframe in increasing value
    Add timeframe to dataframe
    
    """
    
    dataframe[month_year_column] = pd.to_datetime(dataframe[date_column]).dt.to_period('M')
    months = list(dataframe[month_year_column].sort_values(ascending = True).astype(str).unique())
    month_value = [i for i in range(1, len(months)+1)]
    months_dict = dict(zip(months, month_value))
    dataframe[new_column] = dataframe[month_year_column].astype(str).map(months_dict)
    return dataframe.head(10)

#Apply timeframe function on month_year column
timeframe(df_new, 'date', 'month_year', 'timeframe')
```

## Define function to calculate metrics to use

```python
def ratio(dataframe, numerator_column, denominator_column, ratio_column):
    
    """
    Function to calculate metrics/ratios to use
    Give the two columns required to calculate ratio and add it as a column to dataframe
    Scale the ratios calculated using StandardScaler
    
    """
    
    dataframe[ratio_column] = dataframe[numerator_column]/dataframe[denominator_column]
    return dataframe.head(10)

#Apply ratio function to calculate P/E ration, entriprise value to ebitda, enterprise value to sales
ratio(df_new, 'price', 'epsNtm', 'P/E')
ratio(df_new, 'entrVal', 'ebitdaNtm', 'Enterprise_to_ebitda')
ratio(df_new, 'entrVal', 'salesNtm', 'Enterprise_to_sales')
```

## Define function for making Histogram and Pivot of Binned Returns

```python
def make_histogram(ratio_column, ratio_threshold, timeframe_startpoint, timeframe_endpoint):
    '''
    Function to make Histogram and Pivot of Binned Returns, based on Ratio, threshold and time period.
    First Take a subset of dataframe based on threshold and timeframe
    Make a pivot table of the binned return frequencies(Binned Price Changes)
    Make Histogram of Price Change for that Time Period
    
    '''
    
    df_binned_return = df_new[df_new[ratio_column] >= ratio_threshold][(df_new['timeframe'] > timeframe_startpoint) &
                                                                            (df_new['timeframe'] <= timeframe_endpoint)]
    pivot_binned_return = pd.pivot_table(df_binned_return, index = 'Price Change Interval',
                                         values = ratio_column, fill_value = 0,
                                         columns = 'timeframe', aggfunc = 'count')
    print(pivot_binned_return)
    
    plt.figure(figsize = (10, 10))
    sns.distplot(df_binned_return['Price Change'], label = 'Price Changes', kde = False, color = 'coral')
    plt.legend()
    plt.xlabel('Price Changes Distribution')
    plt.ylabel('Frequencies')
    plt.xticks(np.arange(0, 1001, 200))
    plt.title('''Distribution of Price Changes with {} ratio greater than {} over 
              a {} month period.'''.format(ratio_column, ratio_threshold, (timeframe_endpoint - timeframe_startpoint)))
    plt.show()
```

## User Input for Making Histogram

```python
if __name__ == '__main__':
    ratio_column = str(input('Enter Ratio: '))

    ratio_threshold = float(input('Enter ratio threshold: '))

    timeframe_startpoint = int(input('Enter start of time period: '))
    
    timeframe_endpoint = int(input('Enter end of time period: '))

    make_histogram(ratio_column, ratio_threshold, timeframe_startpoint, timeframe_endpoint)
```
